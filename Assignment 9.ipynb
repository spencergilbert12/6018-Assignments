```python
# An introduction to the library and its sources as comments.

#Seaborn is a python library used for data visualization. Seaborn is based off of matplotlib, and performs statistical aggregations to make plots, 
#graphs, and charts to represent data in a useful way. Because seaborn works with panda dataframes, it is easy to create visuals to help understand
#the data. I found and read documentation for seaborn at seaborn.pydata.org.
```

```python
# Advantages, limitations of using the library also provided as comments.

#Pros
#Seaborn is easy to use and requires a minimum amount of coding.
#Seaborn offers a variety of styles and colors for chart/graphs/plots, allowing for good looking visuals for presentations.
#Seaborn is integrated with panda dataframes, making it easy to create visuals directly from the data analysis workflow.

#Cons
#Because seaborn is easy to use and requires a minimum amount of coding, it also isn't as customizable as matplotlib.
#Because seaborn creates a high-level interface, it can be difficult to debug when a problem with a plot occurs.
```

```python
# Code to demonstrate calling the library, and use an example dataset to run the library.

import pandas as pd                                     #Will need pandas to read the dataset
import seaborn as sns                                   #Import Seaborn

sns.set_theme()                                         #Set default seaborn theme for graphs

diabetes_data = pd.read_csv('diabetic_data.csv')        #Use pandas to read dataset csv file

g = sns.catplot(                                        #Create a catplot
    data = diabetes_data, kind = "bar",                 #Use the diabetes dataset and create bars for graphical representation
    x = "age", y = "num_medications", hue = "gender",   #Set columns in dataset as x axis, y axis, and additional variable (hue)
    errorbar="sd", palette="dark", alpha=.6, height=6   #Dimensions for bars
) 
g.despine(left=True)                                    #Format chart
g.set_axis_labels("Age Range", "# of Medications")      #Labels for x and y axis
g.legend.set_title("Gender")                            #Label legend
```

```python
sns.displot(                                                         #Create a displot in seaborn 
    diabetes_data, x="time_in_hospital", col="age", row="gender",    #Set variables, x-axis, column, and row
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),        #Set dimensions for chart
)
```
